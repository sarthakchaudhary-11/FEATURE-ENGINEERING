```python
# Data interpolation >> process of estimating unknown values from known values.

#Linear interpolation
#cubic interpolation
#polynomial interpolation
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([1,3,5,7,9])

plt.scatter(x,y)
plt.show()
```

```python
#linear interpolation >> Linear in nature
x_new = np.linspace(1,5,15)
y_interp = np.interp(x_new , x , y)
y_interp
```

```python
plt.scatter(x_new,y_interp)
```

```python
#cubic interpolation
x = np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125])

plt.scatter(x,y)
plt.show()
```

```python
from scipy.interpolate import CubicSpline
cs = CubicSpline(x,y)
x_new = np.linspace(1,5,15)
y_interp = cs(x_new)
y_interp
```

```python
plt.scatter(x_new,y_interp)
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,4,16,25,6])

plt.scatter(x,y)
plt.show()
```

```python
x_new = np.linspace(1,5,15)
y_interp = np.polyval(np.polyfit(x,y,10),x_new)
y_interp
```

```python
plt.scatter(x_new , y_interp)
```
